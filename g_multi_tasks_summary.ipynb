```python
max_tasks_per_env=5
max_tasks_composed=10
np.random.seed(42)
task_configs={}

for scene_name in np.sort(list(scene_dict.keys())):
    with open(os.path.join(root_path,'../../multiagent_longtasks/{}.json'.format(scene_name)),'r') as f :
        task_config=json.load(f)
    for n_task in range(1,max_tasks_composed):
        if "tasks_{}".format(n_task) in task_config.keys():
            task_lists=task_config["tasks_{}".format(n_task)]
            for idx in range(len(task_lists)):
                if "finetuned" in task_lists[idx].keys():
                    for id,compose in enumerate(np.sort(list(task_lists[idx]["finetuned"].keys()))):
                        task_configs["{}|{}|compose_{}|fintune_{}".format(scene_name,"tasks_{}".format(n_task),idx,id)]=task_lists[idx]["finetuned"][compose]
                else:
                    task_configs["{}|{}|compose_{}|fintune_1".format(scene_name,"tasks_{}".format(n_task),idx)]=task_lists[idx]["task_list"][0]
            
with open(os.path.join(root_path,'../../multiagent_longtasks_summary.json'),'w') as f :
    f.write(json.dumps(task_configs,sort_keys=False,indent=4,separators=(',',':')))
```
